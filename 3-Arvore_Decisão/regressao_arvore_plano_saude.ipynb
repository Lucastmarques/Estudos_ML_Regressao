# Árvore de Decisão para Regressão

O que precisamos ter em mente nesse modelo é que a árvore de decisão não é um modelo contínuo, ou seja, ele irá sempre retornar valores discretos. Além disso, isso faz com que a árvore de decisão também não seja um modelo linear e sim uma "escada", cortando os dados em pontos específicos até conseguir separá-los adequadamente.

```python
import pandas as pd 

base = pd.read_csv('plano_saude2.csv')

X = base.iloc[:, 0:1].values
y = base.iloc[:, 1].values
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='mse')
regressor.fit(X, y)
```

```python
score = regressor.score(X, y)
score
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.plot(X, regressor.predict(X), color = 'red')
plt.title('Regressão com Árvores')
plt.xlabel('Idade')
plt.ylabel('Custo')
```

```python
import numpy as np 
X_teste = np.arange(min(X), max(X)+0.1, 0.1)
X_teste = X_teste.reshape(-1,1)

plt.scatter(X, y)
plt.plot(X_teste, regressor.predict(X_teste), color = 'red')
plt.title('Regressão com Árvores')
plt.xlabel('Idade')
plt.ylabel('Custo')
```
